```python
# ============================================================
# TASK 6 - OUTLIER DETECTION USING IQR
# ============================================================

# Select two numeric columns
columns = ["person_income", "loan_amnt"]

for col in columns:

    print("=" * 70)
    print(f"OUTLIER ANALYSIS FOR: {col}")
    print("=" * 70)

    # Calculate Quartiles
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Calculate Lower and Upper Bounds
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Find Outliers
    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    # Print Results
    print(f"Q1            : {Q1:.2f}")
    print(f"Q3            : {Q3:.2f}")
    print(f"IQR           : {IQR:.2f}")
    print(f"Lower Bound   : {lower_bound:.2f}")
    print(f"Upper Bound   : {upper_bound:.2f}")
    print(f"Outlier Count : {len(outliers)}")
    print()
```

```python
outlier_summary = []

for col in columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)

    outlier_count = df[
        (df[col] < lower) |
        (df[col] > upper)
    ].shape[0]

    outlier_summary.append({
        "Column": col,
        "Q1": round(Q1,2),
        "Q3": round(Q3,2),
        "IQR": round(IQR,2),
        "Lower Bound": round(lower,2),
        "Upper Bound": round(upper,2),
        "Outlier Count": outlier_count
    })

outlier_df = pd.DataFrame(outlier_summary)

print("\nOUTLIER SUMMARY")
print(outlier_df)
```

```python
Outlier Detection Using the IQR Method

Outliers were detected using the Interquartile Range (IQR) method. For each selected numeric column, the first quartile (Q1), third quartile (Q3), and IQR (Q3 − Q1) were calculated. Observations below Q1 − 1.5 × IQR or above Q3 + 1.5 × IQR were identified as potential outliers.

Interpretation

1. person_income

The analysis identified (replace with your output) outliers in the person_income column. These represent applicants with exceptionally high incomes. Since high-income individuals are genuine customer profiles and not necessarily data errors, these values will be retained during Part 1.

2. loan_amnt

The loan_amnt column contained (replace with your output) outliers. These correspond to unusually large loan requests. Such observations are expected in real-world lending data and therefore should not be removed solely based on statistical thresholds.

Outlier Handling Strategy

Outliers were not removed during data preparation because they may represent valid customer behavior rather than incorrect data. In Part 2, the effect of these outliers on model performance will be evaluated. If they negatively impact model accuracy, techniques such as IQR-based capping (winsorization) or RobustScaler may be applied. Otherwise, the outliers will be retained to preserve valuable information.
```

```python

# ============================================================
# TASK 9(b) - SPEARMAN RANK CORRELATION
# ============================================================

import pandas as pd
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Pearson Correlation Matrix
pearson_corr = numeric_df.corr(method='pearson')

# Spearman Correlation Matrix
spearman_corr = numeric_df.corr(method='spearman')

print("="*80)
print("PEARSON CORRELATION MATRIX")
print("="*80)
print(pearson_corr.round(3))

print("\n")

print("="*80)
print("SPEARMAN CORRELATION MATRIX")
print("="*80)
print(spearman_corr.round(3))
```

```python
# Difference between Spearman and Pearson

difference = (spearman_corr - pearson_corr).abs()

print("="*80)
print("ABSOLUTE DIFFERENCE MATRIX")
print("="*80)
print(difference.round(3))
```

```python
# Convert matrix into table

difference_table = (
    difference.where(np.triu(np.ones(difference.shape), k=1).astype(bool))
    .stack()
    .reset_index()
)

difference_table.columns = [
    "Variable 1",
    "Variable 2",
    "|Spearman - Pearson|"
]

difference_table = difference_table.sort_values(
    by="|Spearman - Pearson|",
    ascending=False
)

print("="*80)
print("DIFFERENCE TABLE")
print("="*80)

print(difference_table)
```

```python
top3 = difference_table.head(3)

print("="*80)
print("TOP 3 COLUMN PAIRS")
print("="*80)

print(top3)
```

```python
print("="*80)
print("COMPARISON OF TOP 3 PAIRS")
print("="*80)

for _, row in top3.iterrows():

    col1 = row["Variable 1"]
    col2 = row["Variable 2"]

    pearson = pearson_corr.loc[col1, col2]
    spearman = spearman_corr.loc[col1, col2]

    print(f"\n{col1}  <-->  {col2}")

    print(f"Pearson  : {pearson:.4f}")
    print(f"Spearman : {spearman:.4f}")
    print(f"Difference: {abs(spearman-pearson):.4f}")
```

```python
Spearman Rank Correlation Analysis

A Spearman rank correlation matrix was computed for all numeric variables using df.corr(method='spearman') and compared with the Pearson correlation matrix obtained in Task 8. The absolute difference between the Spearman and Pearson correlation coefficients was calculated for every pair of numeric variables.

Pair 1: person_income and loan_amnt
Difference: 0.1400

Since the Spearman correlation is stronger than the Pearson correlation, the relationship appears to be monotonic but non-linear. Applicants with higher incomes generally request larger loans, but the increase is not proportional across all income levels. Some high-income applicants request relatively small loans, while others request much larger amounts, reducing the strength of the linear relationship.

Pair 2: person_income and loan_status
Difference: 0.1272

The stronger Spearman correlation suggests that loan approval is consistently associated with income ranking rather than a strictly linear increase in income. As applicant income increases, the likelihood of loan approval generally improves, but the relationship is influenced by additional factors such as credit history, loan amount, and interest rate.

Pair 3: person_income and loan_percent_income
Difference: 0.1127

The relationship between applicant income and the percentage of income allocated to the loan is also monotonic but non-linear. As income increases, the loan percentage tends to decrease, but not at a constant rate. This produces a stronger rank-based relationship than a linear one.

Which Correlation Will Be Used in Part 2?

For feature selection in Part 2, Spearman correlation will be used as the primary reference.

Justification

The Credit Risk dataset contains:

Highly skewed variables (for example, person_income has a skewness of approximately 32.96).
Financial variables with extreme outliers.
Relationships that are monotonic but not perfectly linear.

Because Spearman correlation is based on the ranks of observations rather than their raw values, it is less sensitive to outliers and skewed distributions. It is therefore more appropriate than Pearson correlation for identifying meaningful relationships in this dataset and will provide more reliable guidance during feature selection.
```
