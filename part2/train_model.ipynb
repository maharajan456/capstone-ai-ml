````python
# ============================================================
# PART 2 - TASK 1
# Load the Cleaned Dataset
# ============================================================

import pandas as pd

# Load cleaned dataset
df = pd.read_csv("cleaned_data.csv")

print("=" * 60)
print("Dataset Loaded Successfully")
print("=" * 60)

print("\nShape of Dataset:")
print(df.shape)

print("\nFirst Five Rows:")
print(df.head())
````

````python
# Regression Target
y_reg = df["loan_amnt"]

print("\nRegression Label (y_reg): loan_amnt")
print(y_reg.head())
````

````python
# Classification Target
y_clf = df["loan_status"]

print("\nClassification Label (y_clf): loan_status")
print(y_clf.value_counts())
````

````python
#Remove both target columns from the feature matrix.
````

````python
# Feature Matrix

X = df.drop(columns=["loan_status", "loan_amnt"])

print("\nFeature Matrix Shape:")
print(X.shape)

print("\nFeature Columns:")
print(X.columns.tolist())
````

````python
print("=" * 60)
print("SUMMARY")
print("=" * 60)

print("Feature Matrix (X) Shape :", X.shape)
print("Regression Target Shape  :", y_reg.shape)
print("Classification Target Shape :", y_clf.shape)
````

````python
Feature Matrix (X)

The feature matrix contains all predictor variables except the target variables used for regression and classification.

Excluded columns:

loan_amnt
loan_status
Regression Target (y_reg)

The regression target is loan_amnt, a continuous numeric variable representing the amount of money requested by an applicant. This target is appropriate for regression because it contains continuous values.

Classification Target (y_clf)

The classification target is loan_status, an existing binary variable in the dataset.

Typical values are:

0 → Loan Not Defaulted (or Approved)
1 → Loan Defaulted (or Rejected)

Since the dataset already provides a natural binary target, no additional binarization was required.

````

````python
# ============================================================
# STEP 1 - LABEL ENCODE ORDINAL COLUMN
# ============================================================

# Create an ordered mapping for loan_grade
grade_mapping = {
    'A': 1,
    'B': 2,
    'C': 3,
    'D': 4,
    'E': 5,
    'F': 6,
    'G': 7
}

# Apply mapping
X["loan_grade"] = X["loan_grade"].map(grade_mapping)

print("Loan Grade Encoding Completed")

print(X["loan_grade"].head())
````

````python
# ============================================================
# STEP 2 - ONE-HOT ENCODING
# ============================================================

nominal_columns = [
    "person_home_ownership",
    "loan_intent",
    "cb_person_default_on_file"
]

X = pd.get_dummies(
    X,
    columns=nominal_columns,
    drop_first=True
)

print("=" * 60)
print("One-Hot Encoding Completed")
print("=" * 60)

print("Shape after Encoding:")
print(X.shape)
````

````python
print("=" * 60)
print("Encoded Feature Matrix")
print("=" * 60)

print(X.head())
````

````python
print("=" * 60)
print("Data Types After Encoding")
print("=" * 60)

print(X.dtypes)
````

````python
Categorical Feature Encoding

The feature matrix contained four categorical variables.

Ordinal Encoding

The loan_grade column was encoded using manual label encoding because its categories have a meaningful order representing increasing credit risk.

Grade	Encoded Value
A	1
B	2
C	3
D	4
E	5
F	6
G	7

This encoding preserves the natural ordering of loan grades while allowing machine learning algorithms to process the values numerically.

One-Hot Encoding

The remaining categorical variables:

person_home_ownership
loan_intent
cb_person_default_on_file

do not have a natural ordering. They were therefore encoded using One-Hot Encoding with drop_first=True.

Using one-hot encoding creates separate binary indicator columns for each category. Setting drop_first=True removes one dummy variable from each categorical feature, reducing multicollinearity (the dummy variable trap) while retaining all necessary information.

Unlike label encoding, one-hot encoding does not introduce an artificial ordinal relationship between categories. For example, assigning integers to home ownership categories (e.g., RENT = 1, OWN = 2, MORTGAGE = 3) would incorrectly imply that these categories have an inherent ranking, which is not true. One-hot encoding avoids this issue by representing each category independently.
````

````python
# ============================================================
# PART 2 - TASK 3
# Leak-Free Train-Test Split and Feature Scaling
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# ============================================================
# TRAIN-TEST SPLIT FOR REGRESSION
# ============================================================

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X,
    y_reg,
    test_size=0.20,
    random_state=42
)

print("=" * 60)
print("Regression Train-Test Split")
print("=" * 60)

print("X_train_reg :", X_train_reg.shape)
print("X_test_reg  :", X_test_reg.shape)
print("y_train_reg :", y_train_reg.shape)
print("y_test_reg  :", y_test_reg.shape)
````

````python
# ============================================================
# TRAIN-TEST SPLIT FOR CLASSIFICATION
# ============================================================

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X,
    y_clf,
    test_size=0.20,
    random_state=42
)

print("=" * 60)
print("Classification Train-Test Split")
print("=" * 60)

print("X_train_clf :", X_train_clf.shape)
print("X_test_clf  :", X_test_clf.shape)
print("y_train_clf :", y_train_clf.shape)
print("y_test_clf  :", y_test_clf.shape)
````

````python
# ============================================================
# STANDARD SCALING FOR REGRESSION
# ============================================================

scaler_reg = StandardScaler()

# Fit only on training data
scaler_reg.fit(X_train_reg)

# Transform training and testing data
X_train_reg_scaled = scaler_reg.transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

print("=" * 60)
print("Regression Feature Scaling Completed")
print("=" * 60)

print("Scaled Training Shape :", X_train_reg_scaled.shape)
print("Scaled Testing Shape  :", X_test_reg_scaled.shape)
````

````python
# ============================================================
# STANDARD SCALING FOR CLASSIFICATION
# ============================================================

scaler_clf = StandardScaler()

# Fit only on training data
scaler_clf.fit(X_train_clf)

# Transform training and testing data
X_train_clf_scaled = scaler_clf.transform(X_train_clf)
X_test_clf_scaled = scaler_clf.transform(X_test_clf)

print("=" * 60)
print("Classification Feature Scaling Completed")
print("=" * 60)

print("Scaled Training Shape :", X_train_clf_scaled.shape)
print("Scaled Testing Shape  :", X_test_clf_scaled.shape)
````

````python
# Convert scaled arrays back to DataFrames (optional, for easier inspection)

import pandas as pd

X_train_reg_scaled = pd.DataFrame(
    X_train_reg_scaled,
    columns=X_train_reg.columns
)

X_test_reg_scaled = pd.DataFrame(
    X_test_reg_scaled,
    columns=X_test_reg.columns
)

print("=" * 60)
print("First Five Rows of Scaled Training Data")
print("=" * 60)

print(X_train_reg_scaled.head())
````

````python
Why Was the Scaler Fitted Only on the Training Data?

Fitting the scaler on the entire dataset before splitting would introduce data leakage because the calculated mean and standard deviation would include information from the test set. This allows information from the unseen test data to influence the preprocessing of the training data, leading to overly optimistic model performance estimates.

By fitting the scaler only on the training data, the test set remains completely unseen during preprocessing. This simulates how the model would perform on new, real-world data and provides a fair and unbiased evaluation of model performance.
````

````python
# ============================================================
# PART 2 - TASK 4
# LINEAR REGRESSION
# ============================================================

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np

# ============================================================
# TRAIN LINEAR REGRESSION MODEL
# ============================================================

linear_model = LinearRegression()

linear_model.fit(X_train_reg_scaled, y_train_reg)

print("=" * 60)
print("Linear Regression Model Trained Successfully")
print("=" * 60)
````

````python
# ============================================================
# PREDICTIONS
# ============================================================

y_pred_reg = linear_model.predict(X_test_reg_scaled)

print("Prediction Completed")
````

````python
# ============================================================
# MODEL EVALUATION
# ============================================================

mse_lr = mean_squared_error(y_test_reg, y_pred_reg)

r2_lr = r2_score(y_test_reg, y_pred_reg)

print("=" * 60)
print("LINEAR REGRESSION PERFORMANCE")
print("=" * 60)

print(f"MSE : {mse_lr:.4f}")
print(f"R²  : {r2_lr:.4f}")
````

````python
# ============================================================
# RIDGE REGRESSION
# ============================================================

ridge_model = Ridge(alpha=1.0)

ridge_model.fit(
    X_train_reg_scaled,
    y_train_reg
)

ridge_predictions = ridge_model.predict(
    X_test_reg_scaled
)

ridge_mse = mean_squared_error(
    y_test_reg,
    ridge_predictions
)

ridge_r2 = r2_score(
    y_test_reg,
    ridge_predictions
)

print("=" * 60)
print("RIDGE REGRESSION PERFORMANCE")
print("=" * 60)

print(f"MSE : {ridge_mse:.4f}")
print(f"R²  : {ridge_r2:.4f}")
````

````python
# ============================================================
# MODEL COMPARISON
# ============================================================

comparison = pd.DataFrame({

    "Model":[
        "Linear Regression",
        "Ridge Regression"
    ],

    "MSE":[
        mse_lr,
        ridge_mse
    ],

    "R²":[
        r2_lr,
        ridge_r2
    ]

})

print("=" * 60)
print("MODEL COMPARISON")
print("=" * 60)

print(comparison)
````

````python
Both Linear Regression and Ridge Regression achieved almost identical performance on the test dataset. This indicates that the dataset does not suffer from severe multicollinearity and that the ordinary least squares model already generalizes reasonably well.

Ridge Regression applies L2 regularization, which shrinks coefficient values to reduce overfitting. With alpha = 1.0, the regularization effect is relatively small, resulting in performance that is very similar to Linear Regression. Although Ridge did not improve the evaluation metrics in this case, it remains a valuable technique because it can produce more stable models when predictors are highly correlated.
````

````python


# Performance Comparison

| Model                          |  Precision |     Recall |   F1 Score |        AUC |
| ------------------------------ | ---------: | ---------: | ---------: | ---------: |
| Logistic Regression (C = 1.0)  | **0.5118** | **0.7824** | **0.6188** | **0.8603** |
| Logistic Regression (C = 0.01) | **0.5126** | **0.7803** | **0.6187** | **0.8600** |

---

# Metric-by-Metric Interpretation

### Precision

Baseline:

```text
0.5118
```

Regularized:

```text
0.5126
```

Precision increased by only:

```text
0.00077
```

This improvement is extremely small (less than **0.1%**) and is practically negligible.

---

### Recall

Baseline:

```text
0.7824
```

Regularized:

```text
0.7803
```

Recall decreased by:

```text
0.0021
```

Again, this is a very small decrease.

---

### F1 Score

Baseline:

```text
0.6188
```

Regularized:

```text
0.6187
```

The F1-score is almost unchanged.

---

### AUC

Baseline:

```text
0.8603
```

Regularized:

```text
0.8600
```

The AUC decreased by only:

```text
0.00026
```

This is effectively the same performance.

---

# Did Reducing `C` Improve the Model?

The answer is:

> **No significant improvement was observed.**

The stronger regularization (`C = 0.01`) produced performance that was almost identical to the baseline (`C = 1.0`).

This suggests that:

* The original model was **not overfitting**.
* The dataset is already well-behaved after preprocessing.
* Applying stronger regularization had little impact on the model's ability to generalize.

---

# Why Did This Happen?

Recall that in Logistic Regression:


Smaller C
        ↓
Stronger Regularization
        ↓
Smaller Coefficients
        ↓
Simpler Decision Boundary
```

Your dataset already has:

* Proper feature scaling
* Appropriate encoding
* Balanced training data using SMOTE
* A reasonable number of informative features

Therefore, the baseline Logistic Regression did not require much additional regularization.

---



## Regularization Experiment

A second Logistic Regression model was trained using **`C = 0.01`**, which applies stronger L2 regularization than the baseline model (`C = 1.0`).

The comparison of both models is shown below.

| Model                            |  Precision |     Recall |   F1 Score |        AUC |
| -------------------------------- | ---------: | ---------: | ---------: | ---------: |
| Logistic Regression (`C = 1.0`)  | **0.5118** | **0.7824** | **0.6188** | **0.8603** |
| Logistic Regression (`C = 0.01`) | **0.5126** | **0.7803** | **0.6187** | **0.8600** |

### What Does `C` Control?

The parameter **`C`** controls the strength of L2 regularization in Logistic Regression and is the inverse of the regularization strength.

* A **larger value of `C`** applies weaker regularization, allowing the model to fit the training data more closely.
* A **smaller value of `C`** applies stronger regularization, shrinking the model coefficients toward zero and reducing model complexity.

### Interpretation

Reducing `C` from **1.0** to **0.01** produced only minimal changes in Precision, Recall, F1-score, and AUC. Although Precision increased slightly, Recall, F1-score, and AUC decreased by very small amounts. These differences are negligible and indicate that the baseline Logistic Regression model was already appropriately regularized. Therefore, stronger regularization did not provide a meaningful improvement in predictive performance for this dataset.


````

````python
# ============================================================
# BOOTSTRAP CONFIDENCE INTERVAL FOR AUC DIFFERENCE
# ============================================================

import numpy as np
from sklearn.metrics import roc_auc_score
# Reproducibility

np.random.seed(42)
# ============================================================
# BOOTSTRAP SAMPLING
# ============================================================

n_bootstrap = 500

auc_differences = []

for i in range(n_bootstrap):

    # Sample indices with replacement
    sample_index = np.random.choice(
        len(y_test_clf),
        size=len(y_test_clf),
        replace=True
    )

    # Bootstrap sample
    y_boot = y_test_clf.iloc[sample_index]

    prob_c1 = y_probability[sample_index]

    prob_c001 = y_prob_reg[sample_index]

    # Skip samples containing only one class
    if len(np.unique(y_boot)) < 2:
        continue

    auc1 = roc_auc_score(
        y_boot,
        prob_c1
    )

    auc2 = roc_auc_score(
        y_boot,
        prob_c001
    )

    auc_differences.append(
        auc1 - auc2
    )
    
````

````python
# ============================================================
# CONFIDENCE INTERVAL
# ============================================================

mean_difference = np.mean(auc_differences)

lower = np.percentile(
    auc_differences,
    2.5
)

upper = np.percentile(
    auc_differences,
    97.5
)

print("=" * 60)
print("BOOTSTRAP RESULTS")
print("=" * 60)

print(f"Mean AUC Difference : {mean_difference:.6f}")
print(f"95% CI Lower Bound  : {lower:.6f}")
print(f"95% CI Upper Bound  : {upper:.6f}")
````

````python

## Bootstrap Confidence Interval for AUC Difference

To determine whether the baseline Logistic Regression model (`C = 1.0`) consistently outperformed the strongly regularized model (`C = 0.01`), a bootstrap analysis was performed.

A total of **500 bootstrap samples** were generated from the test dataset by sampling observations with replacement. For each bootstrap sample:

* The AUC of the baseline Logistic Regression model (`C = 1.0`) was calculated.
* The AUC of the regularized Logistic Regression model (`C = 0.01`) was calculated.
* The difference in AUC (`AUC(C=1.0) − AUC(C=0.01)`) was recorded.

The mean AUC difference and the 95% confidence interval were then computed from the 500 bootstrap samples.

### Bootstrap Results

| Metric                                |         Value |
| ------------------------------------- | ------------: |
| Mean AUC Difference                   |  **0.000283** |
| 95% Confidence Interval (Lower Bound) | **-0.000127** |
| 95% Confidence Interval (Upper Bound) |  **0.000666** |

### Interpretation

The mean AUC difference of **0.000283** indicates that, on average, the baseline Logistic Regression model (`C = 1.0`) achieved a slightly higher AUC than the model with stronger regularization (`C = 0.01`). However, the improvement is extremely small (approximately **0.03%**).

The **95% confidence interval ranges from -0.000127 to 0.000666**. Since this interval **includes zero**, there is insufficient statistical evidence to conclude that the baseline model consistently outperforms the regularized model across different samples of the data.

This means that the observed difference in AUC is likely due to normal sampling variability rather than a genuine improvement in predictive performance.

### Conclusion

Although the Logistic Regression model with `C = 1.0` achieved a marginally higher AUC than the model with `C = 0.01`, the bootstrap confidence interval indicates that this advantage is **not statistically significant**. Therefore, both models demonstrate essentially the same classification performance on this dataset.

From a practical perspective, either model could be selected. Since the baseline model (`C = 1.0`) provides a slightly higher AUC while maintaining similar precision, recall, and F1-score, it was chosen as the final Logistic Regression model for the remainder of the project.
````
